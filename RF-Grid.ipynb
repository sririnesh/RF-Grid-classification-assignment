```python

import pandas as pd
```

```python
dataset=pd.read_csv("insurance.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
indep=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes','region_northwest', 'region_southeast', 'region_southwest']]
indep
```

```python
dep=dataset[["expenses"]]
```

```python
dep
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import GridSearchCV

param_grid={'criterion':['squared_error','absolute_error','poisson'],'max_features':['sqrt','log2'],
              'n_estimators':[10,100]}

grid= GridSearchCV(RandomForestRegressor(), param_grid, refit=True, verbose=3,n_jobs=-1)

grid.fit(indep,dep)
                      
```

```python
re=grid.cv_results_

print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
region_northwest_input=(input("region northwest 0 or 1:"))
region_southeast_input=(input("region southeast 0 or 1:"))
region_southwest_input=(input("region southeast 0 or 1:"))
```

```python
Future_prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input,region_northwest_input,region_southeast_input,region_southwest_input]])
print("Future_prediction={}".format(Future_prediction))
```
